```python
#2A Perceptron is the simplest type of artificial neural network, functioning as a linear binary classifier. It mimics a single biological neuron by receiving numerical inputs, multiplying them by specific weights, summing them up, and passing the total through an activation function to produce an output.
Key LimitationsLinear Separability Only: A single perceptron can only solve problems where the data can be perfectly separated by a straight line or a flat hyperplane. It fundamentally fails on non-linear data.The XOR Problem: Because of its linear nature, a basic perceptron cannot learn non-linear patterns like the logical XOR (Exclusive OR) operation.Binary Outputs Only: It is restricted to binary classification tasks (e.g., Yes/No, 0/1) and cannot output probabilities or handle multi-class classification natively.
```

```python
#3An activation function determines whether a neuron fires by introducing non-linearity into the network. Without it, the neural network acts as a single linear regression model, incapable of learning complex patterns like image recognition or natural language processing.
```

```python
#5
Optimizers in neural networks iteratively adjust learnable parameters (weights and biases) to minimize the loss function. They dictate how fast and efficiently the model learns from data by dictating how much to update parameters based on the calculated gradients.
```

```python
#6
import numpy as np


class Perceptron:

    def __init__(self, input_size, learning_rate=0.1, epochs=10):
        """Initializes weights, bias, learning rate, and training epochs."""
        # Initialize weights to zeros (or small random numbers)
        self.weights = np.zeros(input_size)
        self.bias = 0.0
        self.lr = learning_rate
        self.epochs = epochs

    def activation_function(self, x):
        """Heaviside step function: returns 1 if input >= 0, else 0."""
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        """Calculates the dot product of inputs and weights, adds bias, and applies activation."""
        linear_output = np.dot(inputs, self.weights) + self.bias
        return self.activation_function(linear_output)

    def train(self, X, y):
        """Trains the perceptron using the perceptron learning rule."""
        for epoch in range(self.epochs):
            total_error = 0
            for inputs, label in zip(X, y):
                prediction = self.predict(inputs)
                error = label - prediction

                if error != 0:
                    # Update weights: w = w + lr * error * input
                    self.weights += self.lr * error * inputs
                    # Update bias: b = b + lr * error
                    self.bias += self.lr * error
                    total_error += abs(error)

            print(f"Epoch {epoch+1}: Total Errors = {total_error}")

            # Stop training early if the model has perfectly converged
            if total_error == 0:
                print("Convergence reached. Stopping training.")
                break
```

```python
#7import numpy as np
import matplotlib.pyplot as plt

# 1. Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# 2. Generate input data range
x = np.linspace(-10, 10, 500)

# 3. Create the plots
plt.figure(figsize=(12, 8))

# Plot Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid(x), color='blue', lw=2, label=r'$\sigma(x) = \frac{1}{1 + e^{-x}}$')
plt.title('Sigmoid Activation Function')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.legend(fontsize=12)

# Plot Tanh
plt.subplot(2, 2, 2)
plt.plot(x, tanh(x), color='red', lw=2, label=r'$tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$')
plt.title('Tanh Activation Function')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.legend(fontsize=12)

# Plot ReLU
plt.subplot(2, 2, 3)
plt.plot(x, relu(x), color='green', lw=2, label=r'$f(x) = \max(0, x)$')
plt.title('ReLU Activation Function')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.legend(fontsize=12)

# Plot All Combined for Comparison
plt.subplot(2, 2, 4)
plt.plot(x, sigmoid(x), color='blue', lw=1.5, label='Sigmoid')
plt.plot(x, tanh(x), color='red', lw=1.5, label='Tanh')
plt.plot(x, relu(x), color='green', lw=1.5, label='ReLU')
plt.title('Comparison View')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.ylim(-1.5, 5.0)  # Constrain y-axis to comfortably see all curves
plt.legend(fontsize=10)

plt.tight_layout()
plt.show()
```
