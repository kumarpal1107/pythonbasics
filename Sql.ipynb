```python
#1
The fundamental difference between DDL, DML, and DQL lies in their operational scope: DDL manages the structure of the database, DML modifies the data content within that structure, and DQL extracts that data without altering it. Think of DDL as building the house, DML as moving or changing the furniture inside, and DQL as looking through the window to see what is inside.
```

```python
#2
SQL constraints are rules enforced on table columns to ensure data accuracy, reliability, and integrity. They automatically prevent invalid data entries, such as missing values or duplicate records, from being saved into the database.Three common types of constraints include:PRIMARY KEY: Uniquely identifies each row in a table. It acts as a distinct identifier (like a row ID) and does not allow NULL values.Scenario: In an Employees table, setting the EmployeeID as a PRIMARY KEY ensures no two employees are assigned the same ID number and every employee has a unique identifier.FOREIGN KEY: Establishes a link between the data in two tables and ensures referential integrity. It prevents actions that would sever the link or create orphaned records.Scenario: In an Orders table, a CustomerID can be set as a FOREIGN KEY that points to the CustomerID in a Customers table. This prevents someone from placing an order for a customer who doesn't exist.CHECK: Ensures that all values in a column satisfy a specific, user-defined condition.Scenario: In a Products table, a CHECK constraint can ensure that the Price column is always greater than \(\$0\), preventing the accidental entry of free or negative-cost items.
```

```python
#3
Key DifferencesLIMIT: Restricts the maximum size of the final result set. For example, LIMIT 10 guarantees you will receive no more than 10 rows.OFFSET: Defines the starting point of the data retrieval by skipping a designated number of rows from the top of the ordered data set.Retrieving the Third Page of ResultsTo calculate pagination parameters programmatically, you can use these standard formulas:\(\text{LIMIT} = \text{Rows Per Page}\)\(\text{OFFSET} = (\text{Page Number} - 1) \times \text{Rows Per Page}\)Given a requirement for Page 3 with 10 records per page:\(\text{LIMIT} = 10\)\(\text{OFFSET} = (3 - 1) \times 10 = 20\)
```

```python
#4
A Common Table Expression (CTE) is a temporary, named result set in SQL that you can reference within a single SELECT, INSERT, UPDATE, or DELETE statement. Defined using the WITH keyword, a CTE acts like a lightweight, virtual table that exists only during the execution of that specific query.Main Benefits of Using a CTEImproves Readability: Breaks monolithic queries into small, logical blocks.Promotes Reusability: References the same temporary result multiple times in one query.Replaces Subqueries: Eliminates deeply nested subqueries that are hard to maintain.Enables Recursion: Allows queries to reference themselves to traverse hierarchical data like org charts.
```

```python
#5
SQL Normalization is a systematic database design technique used to organize data into multiple related tables. The primary goal is to minimize data redundancy and eliminate data anomalies (insertion, update, and deletion errors). By breaking down complex, messy tables into smaller, structured ones, it ensures that data dependencies are logical and that every piece of data is stored in exactly one place.
```
