```python
#1
In a Convolutional Neural Network (CNN), filters (or kernels) act as feature detectors that scan an image to extract patterns like edges and textures. Feature maps are the resulting outputs, acting as "activation maps" that show the exact location and strength of those detected patterns across the image.
```

```python
#2
Padding expands the spatial dimensions of an input image by adding extra rows and columns (usually zeros) around its outer borders, while stride specifies the step size (in pixels) by which the convolutional filter moves across that input. Together with kernel size, these two hyperparameters dictate how aggressively a network extracts information and downsamples data through its layers.
```

```python
#3
 In a Convolutional Neural Network (CNN), the receptive field is the specific region of the original input image that influences the activation of a given feature map pixel. As you go deeper into the network, stacking convolutional layers causes the receptive field to grow exponentially, allowing the model to detect larger, more complex global features.Why It Is Important for Deep ArchitecturesHierarchical Feature Extraction: It enables the network to build features hierarchically. Early layers have small receptive fields, capturing low-level details like edges and colors. Deeper layers aggregate these into larger receptive fields to recognize complex objects (e.g., faces, cars).Global Context: By the final layers, the receptive field should ideally cover the entire input image. This allows the network to make decisions based on the global context of the scene rather than isolated fragments.
```

```python
#4
Filter size directly increases the number of parameters in a convolutional layer, while stride has zero direct impact on it. However, stride indirectly affects the parameter count of downstream Fully Connected (FC) layers by changing the spatial dimensions of the feature maps.
```

```python
#5
Filter Size Shift: LeNet and early AlexNet layers used large receptive fields (like 11x11 in AlexNet) to capture large-scale features. However, VGG discovered that stacking multiple smaller 3x3 filters has the same effective receptive field as one large filter, but with fewer parameters and added non-linearity to learn more complex features.Depth Scaling: The progression from 8 layers (AlexNet) to 16/19 layers (VGG) reflects the deep learning community's realization that increasing depth while shrinking filter sizes significantly boosts the model's ability to identify intricate patterns.Performance: While LeNet was a breakthrough for basic grayscale inputs, it struggles with the high-dimensional variations of modern RGB datasets. AlexNet drastically revolutionized the AI landscape, while VGG fine-tuned this by taking top-3 accuracy honors in 2014 due to its pure, deep structural simplicity.Computational Cost: LeNet is computationally trivial. AlexNet required multi-GPU hardware to parallelize and train. VGG, though performing vastly better, is parameter-heavy and resource-intensive to train and deploy due to its large number of parameters in the fully-connected layers.
```

```python
#6
import os
# Suppress TensorFlow background logging for clean output
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# =====================================================================
# 1. DATA PREPROCESSING
# =====================================================================
# Load the MNIST dataset (60,000 training images, 10,000 test images)
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape images to include the channel dimension (28x28x1 for grayscale)
x_train = x_train.reshape(-1, 28, 28, 1).astype("float32")
x_test = x_test.reshape(-1, 28, 28, 1).astype("float32")

# Normalize pixel values from [0, 255] to the [0.0, 1.0] range
x_train /= 255.0
x_test /= 255.0

print(f"Training data shape: {x_train.shape}")
print(f"Test data shape: {x_test.shape}\n")

# =====================================================================
# 2. MODEL CREATION (CNN Architecture)
# =====================================================================
model = models.Sequential([
    # First Convolutional Block
    layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(pool_size=(2, 2)),

    # Second Convolutional Block
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),

    # Flatten and Dense Classification Layers
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax') # 10 output classes for digits 0-9
])

# Display a summary of the network structure
model.summary()

# =====================================================================
# 3. MODEL COMPILATION
# =====================================================================
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', # Used because labels are integers
    metrics=['accuracy']
)

# =====================================================================
# 4. MODEL TRAINING
# =====================================================================
print("\n--- Training Starting ---")
# Training for 2 epochs for quick demonstration
history = model.fit(
    x_train, y_train,
    epochs=2,
    batch_size=64,
    validation_split=0.1, # Uses 10% of training data to monitor overfitting
    verbose=1
)

# =====================================================================
# 5. MODEL EVALUATION
# =====================================================================
print("\n--- Evaluation Starting ---")
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)

print(f"Final Test Loss: {test_loss:.4f}")
print(f"Final Test Accuracy: {test_acc * 100:.2f}%")
```

```python
#7
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# 1. Load the CIFAR-10 dataset
# The dataset contains 60,000 32x32 color images in 10 classes
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# 2. Preprocess the data
# Normalize pixel values to be between 0 and 1 by dividing by 255.0
train_images, test_images = train_images / 255.0, test_images / 255.0

# Output dataset shapes to verify loading
print(f"Train images shape: {train_images.shape}")
print(f"Test images shape: {test_images.shape}")

# 3. Create the CNN Architecture
model = models.Sequential()

# Convolutional Block 1: Extracts low-level features (edges, textures)
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))

# Convolutional Block 2: Extracts complex features
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Convolutional Block 3: Further feature extraction
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Dense Layers: Classification
model.add(layers.Flatten())  # Convert 3D feature maps to 1D vectors
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10)) # Output layer for 10 classes

# 4. Display the architecture
model.summary()
```
